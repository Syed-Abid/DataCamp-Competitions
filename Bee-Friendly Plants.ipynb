## üìñ Background
You work for the local government environment agency and have taken on a project about creating pollinator bee-friendly spaces. You can use both native and non-native plants to create these spaces and therefore need to ensure that you use the correct plants to optimize the environment for these bees.

## üíæ The Data
You have assembled information on the plants and bees research in a file called `plants_and_bees.csv`. Each row represents a sample that was taken from a patch of land where the plant species were being studied.

| Column     | Description              |
|------------|--------------------------|
| `sample_id` | The ID number of the sample taken. |
| `bees_num` | The total number of bee individuals in the sample. |
| `date` | Date the sample was taken. |
| `season` | Season during sample collection ("early.season" or "late.season"). |
| `site` | Name of collection site. |
| `native_or_non` | Whether the sample was from a native or non-native plot. |
| `sampling` | The sampling method. |
| `plant_species` | The name of the plant species the sample was taken from. None indicates the sample was taken from the air. |
| `time` | The time the sample was taken. |
| `bee_species` | The bee species in the sample. |
| `sex` | The gender of the bee species. |
| `specialized_on` | The plant genus the bee species preferred. |
| `parasitic` | Whether or not the bee is parasitic (0:no, 1:yes). |
| `nesting` | The bees nesting method. |
| `status` | The status of the bee species. |
| `nonnative_bee` | Whether the bee species is native or not (0:no, 1:yes). |

```python
# Start coding here
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv("data/plants_and_bees.csv")
```

```python
#Data Understanding and Cleaning
#print(df)
df.shape
df.head(5)
#df.dtypes
#df.describe()
df.columns
#Now we will be removing/dropping those columns that we don't need
df.drop(['status'],axis=1)
df.drop(['time'],axis=1)
```

```python
#Now we will be renaming columns.
df.rename(columns={'bee_species':'Bee_Species',
                   'sample_id':'Sample_Id',
                   'bees_num':'Bees_Num',
                   'date':'Date',
                   'season':'Season',
                   'sampling':'Sampling',
                   'bee_species':'Bee_Species',
                   'site':'Site',
                   'native_or_non':'Native_or_Non',
                   'sex':'Sex',
                   'specialized_on':'Specialized_On',
                   'parasitic':'Parasitic',
                   'nesting':'nesting',
                   'nonnative_bee':'NonNative_Bee',
                   'status':'Status'})
```

```python
#To find which plant species attract more native bees?
#male_bees = df[df['sex']=='m']
#male_bees
df = df[df['plant_species'].str.contains('None')==False]
bee_plant_counts = df.groupby(['bee_species', 'plant_species']).size().reset_index(name='count')
equivalent_species = bee_plant_counts[bee_plant_counts['count'] > 10]
equivalent_species.sort_values(by=['count'],ascending=False)
#equivalent_species_max = equivalent_species.max()
#equivalent_species_max
#the above code just have to get filtered out for None plant species
```

```python
# Define the data
bee = equivalent_species['bee_species']
plant = equivalent_species['plant_species']
count = equivalent_species['count']

# Perform the visualization
plt.bar(bee,count,color='yellow',width=0.5)
plt.xlabel('Bee Species')
plt.ylabel('Occurences')
plt.title('Bee Occurences')
plt.show()
plt.bar(plant,count,color='black',width=0.5)
plt.xticks(ticks=range(len(plant)), labels=['\n'.join(cat.split()) for cat in plant])
plt.xlabel('Plant Species')
plt.ylabel('Occurences')
plt.title('Plant Occurences')
plt.show()
```

```python
#Now we will find which specialized plant genus attract more bees
df = df[df['specialized_on'].str.contains('null')==False]
specialized_counts = df['specialized_on'].value_counts()
specialized_counts_sorted = specialized_counts.sort_values(ascending=False)
specialized_counts_sorted
```

```python
#Which plant attract more bees(native or not-native)
bee_plant_counts2 = df.groupby(['specialized_on','native_or_non']).size().reset_index(name='count')
equivalent_species2 = bee_plant_counts2[bee_plant_counts2['count'] >= 1]
equivalent_species2.sort_values(by=['count'],ascending=False)
```

So, we can finalize that **"Penstemon"** plant is a bee-friendly plant specie than any other plant specie for Native bees while there is no such friendly plant specie for non-native bees.
